Project Title:- Student Marks Analyzer

Problem statement:- Writing a code that gets marks of students and shows grade on the basis of their avg marks

Logic:-
we will take inputs as marks from user and store in a array form for specific subjects(out of 100) then we will count total sum then avg then we will gave grade on basis of there mark as
A: Above 85
B: 70–85
C: 50–69
Fail: Below 50

1 function to gets marks between 0 and 100
2 inputs
3 sum,avg,highest,lowest
4 grade

```python

print("Welcome to Student Marks Analyzer \n Know your marks fast and accurately\n")

def get_marks(subject):

    while True:
        marks = int(input(f"Enter mark of {subject} Subject : "))

        if marks >=0 and marks <101:
            return marks

        else:
            print("Invalid input! Marks must be between 0 and 100.")


Python = get_marks("Python")
SQL = get_marks("SQL")
Javascript = get_marks("Javascript")
Java = get_marks("Java")
DBMS = get_marks("DBMS")

# !pip install numpy

import numpy as np

marks=np.array([Python,SQL,Javascript,Java,DBMS])


print(f"\nTotal marks :{np.sum(marks)}")
print(f"Average marks :{np.mean(marks)}")
print(f"Highest mark :{np.max(marks)}")
print(f"Lowest mark :{np.min(marks)}")

grade=np.mean(marks)
if grade > 85:
  print("\nGRADE: A")
elif grade >=70 and grade <=85:
  print("GRADE: B")
elif grade >=50 and grade <=69:
  print("GRADE: C")
else:
  print("GRADE: FAIL")
```
